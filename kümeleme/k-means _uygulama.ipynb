# elbow yöntemi

```python
df
```

```python
ssd = []
K = range(1,30)

for k in K:
    kmeans = KMeans(n_clusters = k).fit(df)
    ssd.append(kmeans.inertia_)
#uzaklık farklarının karelerinin toplamını ifade ediyor ssd
```

```python
plt.plot(K, ssd, "bx-")
plt.xlabel("farklı K değerlerine karşılık uzaklık artık toplamları")
plt.title("optimum küme sayısı için Elbow yöntemi")
```

# Hiyeararşik Kümeleme

```python
from scipy.cluster.hierarchy import linkage
```

```python
hc_complete = linkage(df, "complete")
hc_average = linkage(df, "average")
```

```python
from scipy.cluster.hierarchy import dendrogram
```

```python
plt.figure(figsize = (15,10))
plt.title("hiyerarşik kümeleme dendrogramı")
plt.xlabel("gözlem birimleri")
plt.ylabel("uzaklıklar")
dendrogram(hc_complete,
          leaf_font_size = 10);
```

```python
plt.figure(figsize = (10,5))
plt.title("hiyerarşik kümeleme dendrogramı")
plt.xlabel("gözlem birimleri")
plt.ylabel("uzaklıklar")
dendrogram(hc_complete,
           truncate_mode = "lastp",
           p =10,
           show_contracted = True,
           leaf_font_size = 10);
```

```python
plt.figure(figsize = (10,5))
plt.title("hiyerarşik kümeleme dendrogramı")
plt.xlabel("gözlem birimleri")
plt.ylabel("uzaklıklar")
dendrogram(hc_average,
           truncate_mode = "lastp",
           p =10,
           show_contracted = True,
           leaf_font_size = 10);
```

# temel bileşen analizi

```python
df = pd.read_csv("hitters_0/Hitters.csv")
df.dropna(inplace = True)
df = df._get_numeric_data()
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df = StandardScaler().fit_transform(df)
```

```python
df[0:5,0:5]
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
pca_fit = pca.fit_transform(df)
```

```python
bilesen_df = pd.DataFrame(data = pca_fit, columns = ["birinci_bilesen", "ikinci_bilesen"])
```

```python
bilesen_df
```

```python
pca.explained_variance_ratio_
```

```python
pca.components_[1]
```

```python
#optimum bileşen sayısı
pca = PCA().fit(df)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("bileşen sayısı")
plt.ylabel("kümülatif varyans oranı");
```
